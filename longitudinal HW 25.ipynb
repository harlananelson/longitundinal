```sas
data lena;
input response $ group $ count;
datalines;
Yes Lenalidomide 108
No  Lenalidomide 69
Yes Placebo      35
No  Placebo      141
;
run;

ods select CrossTabFreqs PearsonChiSq LRChiSq;
proc freq order=data;
table response*group/chisq relrisk riskdiff expected norow nopercent;
exact chisq;
wieght count;
run;
ods select all;
```
